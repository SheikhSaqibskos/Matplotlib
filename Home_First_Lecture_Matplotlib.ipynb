```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 500)   # 500 values from 0 to 10
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.plot(x, y_sin, label="Sine")
plt.plot(x, y_cos, label="Cosine")

plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Sine & Cosine Waves")
plt.legend()
plt.grid(True)

plt.show()
```

```python
# âœ… 3. Change Line Colors & Styles
plt.plot(x, y_sin, color='red', linestyle='--', linewidth=2, label="Sine")
plt.plot(x, y_cos, color='blue', linestyle='-', linewidth=1, label="Cosine")

# ðŸŽ¨ Common options:
# Colors:
# 'red', 'blue', 'green', 'yellow', 'orange', 'purple', 'black'

# Line styles:
# '-'  solid  
# '--' dashed  
# '-.' dash-dot  
# ':' dotted

# Marker types:
# 'o' circle  
# 's' square  
# '^' triangle  
# '*' star
```

| x values | y values | Works?  |
| -------- | -------- | ------- |

```python
import numpy as np

x = np.linspace(0, 10, 500)   # SAME length
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)   # rows=1, columns=2, plot=1
plt.plot(x, y_sin)
plt.title("Sine Wave")

plt.subplot(1, 2, 2)   # plot=2
plt.plot(x, y_cos)
plt.title("Cosine Wave")

plt.show()
```

**âœ… 8. Customize Axis Range**

```python
# âœ… 8. Customize Axis Range
plt.plot(x, y_sin)
plt.xlim(0, 10)
plt.ylim(-1, 1)
plt.title("Axis Limits")
plt.show()
```

**âœ… 9. Add Text & Annotations**

```python
# âœ… 9. Add Text & Annotations
plt.plot(x, y_sin)
plt.annotate("Peak", xy=(1.57, 1), xytext=(2.5, 1.3),
             arrowprops=dict(arrowstyle="->"))
plt.title("Annotated Sine Wave")
plt.show()
```

| Axes            | `xlim()`, `ylim()`, `xticks()`, `yticks()`        |
| Multiple graphs | `subplot()`, `figure()`                           |
| Styles          | `plt.style.use()`                                 |
